```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
heart_disease= fetch_ucirepo(id=45)
heart_disease
```

```python
X = heart_disease.data.features 
y = heart_disease.data.targets
```

```python
# metadata 
print(heart_disease.metadata) 
  
# variable information 
print(heart_disease.variables) 
```

```python
set(y.num.values)
```

```python
X.isnull().sum()
```

```python
mean_ca= X["ca"].mean()
mean_thal=X["thal"].mean()
X["ca"].fillna(value=mean_ca,inplace=True)
X["thal"].fillna(value=mean_thal,inplace=True)
X.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scale=scaler.fit_transform(X)
scale=pd.DataFrame(scale)
X_train,X_test,Y_train,Y_test=train_test_split(scale,y,random_state=100, test_size=0.2)
X_train.head()
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
preds=model.predict(X_test)
acc=accuracy_score(Y_test,preds)
print(acc)
```
